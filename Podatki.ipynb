```python
import numpy as np
np.set_printoptions(suppress=True)
```

```python
file_object  = open("rna_tissue.tsv", "r")
data = np.loadtxt('rna_tissue.tsv', delimiter="\t", skiprows=1, dtype=np.str)

print("done")
```

```python
print(data)
```

```python
import collections

slovar1 = collections.defaultdict(list)
lastGene=0
for i in data:
    geneId = i[0]
    if(geneId != lastGene):
        slovar1[i[1]].append({"geneId": geneId})
        lastGene = geneId
    slovar1[i[1]].append({i[2]: i[3]})
```

```python
print(slovar1["LAS1L"])
```

```python
slovar2 = collections.defaultdict(list)
deviation=[]
max=0
min=999999
tkivoMax=""
tkivoMin=""
for i in slovar1:
    max=0
    min=9999999999999999999
    tkivoMax=""
    tkivoMin=""
    deviation=[]
    for j in slovar1[i]:
        
        for b,a in j.items():
            if(b=="geneId"):
                continue
            if(float(a)>max):
                max = float(a)
                tkivoMax=b
            if(float(a)<min and float(a)>0):
                min = float(a)
                tkivoMin=b
            deviation.append(float(a))
    print(np.std(deviation))
    print(i,tkivoMax,max,tkivoMin,min)
    slovar2[i].append({"max": max})    
    slovar2[i].append({"min": min})
    slovar2[i].append({"std diviacija": np.std(deviation)})
    slovar2[i].append({"mediana": np.median(deviation)})
    
```

```python
for a in slovar2:
    for b in slovar2[a]:
        for d,e in b.items():
            slovar1[a].append({d: str(e)})    
```
